```python
import pandas as pd
import math
```

```python
df = pd.read_csv("./Data/raw_vaccination/country_vaccinations.csv")
```

```python
countrywise_vaccination=df[df['country']=='United States']
```

```python
countrywise_vaccination.head()
```

```python
countrywise_vaccination['date'] = pd.to_datetime(countrywise_vaccination['date'])
```

```python
countrywise_vaccination.sort_values(by=['date'],inplace =True)
```

```python
countrywise_vaccination.head()
```

```python
countrywise_vaccination.isna().sum()
```

```python
def replace_last_recent(metric):
    metric = metric.tolist()
    if math.isnan(metric[0]):
        prev = 0
    else:
        prev = metric[0]
    for i in range(len(metric)):
        if math.isnan(metric[i]):
            metric[i] = prev
        prev = metric[i]
    return metric
        
```

```python
#replace_last_recent(countrywise_vaccination['total_vaccinations'])
```

```python
countrywise_vaccination['total_vaccinations'] = replace_last_recent(countrywise_vaccination['total_vaccinations'])
countrywise_vaccination['people_vaccinated'] = replace_last_recent(countrywise_vaccination['people_vaccinated'])
countrywise_vaccination['people_fully_vaccinated'] = replace_last_recent(countrywise_vaccination['people_fully_vaccinated'])
countrywise_vaccination['daily_vaccinations_raw'] = replace_last_recent(countrywise_vaccination['daily_vaccinations_raw'])
countrywise_vaccination['daily_vaccinations'] = replace_last_recent(countrywise_vaccination['daily_vaccinations'])
countrywise_vaccination['total_vaccinations_per_hundred'] = replace_last_recent(countrywise_vaccination['total_vaccinations_per_hundred'])
countrywise_vaccination['people_vaccinated_per_hundred'] = replace_last_recent(countrywise_vaccination['people_vaccinated_per_hundred'])
countrywise_vaccination['people_fully_vaccinated_per_hundred'] = replace_last_recent(countrywise_vaccination['people_fully_vaccinated_per_hundred'])
countrywise_vaccination['daily_vaccinations_per_million'] = replace_last_recent(countrywise_vaccination['daily_vaccinations_per_million'])
```
