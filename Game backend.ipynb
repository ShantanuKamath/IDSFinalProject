```python
max_profit = 0
profits = [0,0,0,0]


amzn_stock = df_stock[['AMZN']].values.tolist()
nflx_stock = df_stock[['NFLX']].values.tolist()
ual_stock = df_stock[['UAL']].values.tolist()
twtr_stock = df_stock[['TWTR']].values.tolist()

stocks = df_stock[['AMZN','NFLX','UAL','TWTR']].values.tolist()
time_range = len(amzn_stock)
weights = np.array([0,40,50,0])
minimal_stock_prices = np.array(stocks[0])
min_price_buy = np.sum(minimal_stock_prices*weights)
max_profit = 0

for i in range(1,time_range):
    max_profit = max(max_profit,np.sum(np.divide((np.array(stocks[i])-minimal_stock_prices),minimal_stock_prices)*weights))
    weighted_value = np.sum(np.array(stocks[i])*(weights))
    if weighted_value < min_price_buy:
        min_price_buy = weighted_value
        minimal_stock_prices = np.array(stocks[i])
        
print(max_profit)
print(minimal_stock_prices)
```
